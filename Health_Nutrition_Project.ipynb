```python
from collections import Counter

text = ' '.join(df['description'].dropna().astype(str))
words = text.split()
common_words = Counter(words).most_common(20)
print(common_words)
```

```python
count =df ['data_type'].value_counts()
count.plot(kind='bar')
plt.xlabel("Food Categories")
plt.ylabel("Count of Foods")
plt.title("Distribution of Food Types")
plt.xticks(rotation=0)
plt.show()
```

```python
from collections import Counter

words = " ".join(df['description']).split()
common = Counter(words).most_common(10)

import pandas as pd
common_words = pd.DataFrame(common, columns=['word','count'])

common_words.plot(x='word', y='count', kind='bar')
plt.title("Most Common Words in Food Descriptions")
plt.xticks(rotation=0)
plt.show()

```

```python
df['desc_length'] = df['description'].apply(len)
df['desc_length'].plot(kind='hist', bins=50)
plt.title("Food Description Length")
plt.xlabel("Number of Characters")
plt.show()

```

```python
print("MEAN ", df['desc_length'].mean())
print("MEDIAN ", df['desc_length'].median())
print("MODE ", df['desc_length'].mode())
```

```python
top1 = df['publication_date'].value_counts()
top1 = df['publication_date'].value_counts()

date = top1.index[0]
count = top1.iloc[0]

print(f"Most of the food entries were published on : {date}, total : {count}")
```

```python
top5 = df['data_type'].value_counts().head(5)
top5.plot(kind='pie', autopct='%1.1f%%')
plt.title("Top 5 Food Data Types")
plt.ylabel("")
plt.show()
```

```python
top_foods = df['description'].value_counts().head(10)
top_foods.plot(kind='barh')
plt.title("Top 10 ood Names")
plt.xlabel("Count")
plt.ylabel("Food Name")
plt.show()
```

```python
top5 = df['data_type'].value_counts().head(5)
top5.plot(kind='pie', autopct='%1.1f%%')
plt.title("Overall Distribution of Food Data Types")
plt.ylabel("")
plt.show()
```

```python
stacked = df.groupby(['date', 'data_type']).size().unstack(fill_value=0)

stacked.plot(kind='bar', stacked=True,figsize=(11,5))
plt.title("Date-wise Distribution of Food Data Types")
plt.xlabel("Publication Date")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.legend(title="Data Type")
plt.show()
```
